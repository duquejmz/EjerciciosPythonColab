```python
import pandas as pd

url = 'https://raw.githubusercontent.com/rubisabel00/ciencias-de-datos-Sena/refs/heads/main/DocumentosNB03/clientes.csv'
datasetCSV = pd.read_csv(url, on_bad_lines='skip')
datasetCSV
```

2. Realizar una descripcion estadistica de la informacion encontrada y normalizarla (Eliminar los null).

```python
df_sin_nulos = datasetCSV.dropna()
df_sin_nulos
```

3. Generar una grafica representativa de la informacion obtenida.

```python
df_sin_nulos = datasetCSV.plot()
df_sin_nulos
```

4. Generar dos consultas de agrupamiento de valor agregado a la informacion obtenida.

```python
est_programa = datasetCSV.groupby('Edad').agg({'Edad': 'count'}) 
print(est_programa,'\n')
```

```python
est_edad = datasetCSV.groupby(['Edad', 'Método de pago']).agg({ 'Edad': 'count'})
print(est_edad)
print()
```

5. Generar nuevas columnas segun sean requeridas para completar la informacion obtenida.

```python
datasetCSV['Edad agrupada'] = pd.cut(datasetCSV['Edad'], bins=[0, 18, 35, 50, 65, 100], labels=['0-18', '19-35', '36-50', '51-65', '66+'])
datasetCSV['Compras por categoria'] = pd.cut(datasetCSV['Frecuencia de compra'], bins=[0, 5, 10, 20, 50, 100], labels=['0-5', '6-10', '11-20', '21-50', '51+'])

datasetCSV.head()
```

6. Exportar el trabajo realizado en los ateriores items a un archivo de excel.

```python
datasetCSV.to_excel('processed_clientes.xlsx', index=False)
```
